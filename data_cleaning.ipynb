```python
import pandas as pd 

df = pd.read_excel('Customer Call List.xlsx')
df
```

<h1> DATA CLEANING 

```python
# dropping duplicates
df = df.drop_duplicates()
```

```python
# dropping column can use, inplace = True to save without reassigning variable
df = df.drop(columns='Not_Useful_Column')
```

```python
# cleaning name column, can do lstrip or rstrip for extremes of string
df['Last_Name'] = df['Last_Name'].str.strip('123./')
```

```python
# standardizing phone number format using r.e, removing dash or slash
# regex must be set to True, future version default is False
df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
df['Phone_Number']
```

```python
# changing phone column intp string to apply function 
df['Phone_Number'] = df['Phone_Number'].astype('str')
df['Phone_Number']
```

```python
# using lambda function to standardize phone format
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3]+ '-' + x[3:6] + '-' + x[6:10])
df['Phone_Number']
```

```python
# replacing nulls with blank
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
df['Phone_Number']
```

```python
df
```

```python
# splitting address column
# future update : pat = the delimiter, n = max no. of splits
df[['Street Address','State','Zip']] = df['Address'].str.split(pat=',',n=2,expand=True)
df
```

```python
# standardizing format 

df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Paying Customer']
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df['Do_Not_Contact']
```

```python
# replacing nulls 
df = df.replace('N/a','')
df = df.fillna('') # to replace Nan, because Nan is null not a str
df
```

```python
# removing do not call rows

for x in df.index:
    if df.loc[x,'Do_Not_Contact']=='Y':
        df.drop(x,inplace=True)
df
```

```python
# removing no phone numbers

for x in df.index: # alternative: df = drop.na(subset='Phone Number'), inplace=True)
    if df.loc[x,'Phone_Number']=='':
        df.drop(x,inplace=True)
df
```

```python
# resetting index: drop = true - drops original index

df.reset_index(drop=True)
```
